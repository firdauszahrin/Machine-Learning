```python
id_new = ' '.join(input_data)
print(id_new)
```

```python
model.predict(id_new) #predict the Part of Speech of input using xlnet model
```

```python
#keep the value contain the NOUN or VERB
mod_pos=[]
mod_rs=[]

for i in model.predict(id_new): 
  if (i[1] == 'NOUN' or i[1] == 'VERB'):
    mod_pos.append(i[1])
    mod_rs.append(i[0])
```

```python
#Trigram pattern to check the basic Malay requirement structure: NOUN VERB NOUN
j=0
x=1
limit = len(mod_pos)-3

while j <= limit:
  if (mod_pos[j] == 'NOUN' and mod_pos[j+1] == 'VERB' and mod_pos[j+2] == 'NOUN'):
    print((x),")",mod_rs[j],"(", mod_pos[j],")", mod_rs[j+1],"(", mod_pos[j+1],")", mod_rs[j+2],"(", mod_pos[j+2],")")
    x=x+1
  j=j+1

if j > 0:
  print("This requirement complies with the basic requirement.")
else:
  print("This requirement does NOT comply with the basic requirement. Hence, can't proceed next steps.")
```
