```python
from sklearn import tree, neighbors, svm, naive_bayes
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import cross_val_score

# Load pandas
import pandas as pd

# Load numpy
import numpy as np

path = r'C:\Users\FZ\Dropbox\FYP Master Degree\SRS\wordlist_140417.csv'
words = pd.read_csv(path)

words.head(5)
```

```python
# Create a dataframe with the four feature variables
df = pd.DataFrame(words, columns=words.columns)

words.head(5)
```

```python
#trained data sets : 75%
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75

# View the top 20 rows
df.head(20)
```

```python
# Create two new dataframes, one with the training rows, one with the test rows
train, test = df[df['is_train']==True], df[df['is_train']==False]

# Show the number of observations for the test and training dataframes
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:', len(test))
```

```python
# Create a list of the feature column's names
features = df.columns[:-2]

features
```

```python
# train['CLASS'] contains the string value. Before we can use it,
# we need to convert each string into a numeric value. So, in this case there
# are 2 values, which have been coded as 0 (Y) or 1 (N).
y = pd.factorize(train['CLASS'])[0]

y
```

```python
test_y = pd.factorize(test['CLASS'])[0]

test_y
```

```python
## CHALLENGE - create 5 classifiers
clf = tree.DecisionTreeClassifier()
clf1 = neighbors.KNeighborsClassifier()
clf2 = svm.SVC()
clf3 = naive_bayes.GaussianNB()
clf4 = RandomForestClassifier()

# Train the classifier to take the training features and learn how they relate
# to the training y (train['CLASS'])

#CHALLENGE - train them on our data
clf = clf.fit(train[features], y)
clf1 = clf1.fit(train[features], y)
clf2 = clf2.fit(train[features], y)
clf3 = clf3.fit(train[features], y)
clf4 = clf4.fit(train[features], y)

#CHALLENGE compare their results and print the best one!
prediction = clf.predict(test[features])
prediction1 = clf1.predict(test[features])
prediction2 = clf2.predict(test[features])
prediction3 = clf3.predict(test[features])
prediction4 = clf4.predict(test[features])

prediction
```
